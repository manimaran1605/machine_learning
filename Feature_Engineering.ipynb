```python
#Binning or bucketing
#Suppose if we have continuous variables, we can convert them into binary variables by bucketing. converting them into bins with ranges
#Suppose if you want to represent a range of values from 0 to 10, we can create N= 5 bins, each has an interval of 2.
def binning(x1, x2, x3, x4, x5):
    if x1 + x2 + x3 + x4 + x5 == 1: 
        return 2 * x1 + 4 * x2 + 6 * x3 + 8 * x4 + 10 * x5
    else:
        return print("more than one variable can not be 1")
#x's are the binary variables, only one of them can be one satisfying the constaint sum_{i} xi = 1
print(binning(0,0,0,1,0))
```

```python
#Normalization converts the rage of values between 0 to 1 or -1 to 1
values = np.linspace(0,20, 200)+np.random.rand(200)
def normalization(x):
    return ((x-np.min(x))/(np.max(x)-np.min(x)))
normalized_values = normalization(values)
plt.plot(values)
```

```python
mu, sigma = 0, 0.1 # mean and standard deviation
random_values = np.random.normal(mu, sigma, 1000)
sns.displot(random_values)
print(np.mean(random_values), np.var(random_values))
```

```python
def standardization(x):
    return (x-np.mean(x))/np.sqrt(np.var(x))
standardized_values = standardization(random_values)
sns.displot(standardized_values)
print(np.mean(standardized_values), np.var(standardized_values)) #standardize the mean to 0 and variance to 1
```

```python
#Data Imputation Technique (if some values are missing):
#take the average of a number before missing and after missing and replace the missing number with this average
def data(min,max,N):
    x = np.linspace(min,max,N)
    x[20] = np.nan
    return x
x_range = data(0, 20, 50)
print('data', x_range)
print('20th value is', x_range[20])
```

```python
def impute(data, i):
    impute_value = 0.5*(data[i-1] + data[i+1])
    data[i] = impute_value
    return data
print('Imputed data is ', impute(x_range, 20))
print('20th value is ', impute(x_range,20)[20])
```
