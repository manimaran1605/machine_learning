```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 100)

y = 2 * x**3 - 4 * x**2 + x + np.random.normal(0, 100, size=x.shape)

# Optional: visualize the data
plt.scatter(x, y, color='blue', label='Polynomial data')
plt.title("Data")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
def kernel(z,zi,var):
    return np.exp(-(z-zi)**2/2*var)
```

```python
def nadaraya_watson_estimator(x,xi,yi,var):
    denom = np.zeros(len(xi))
    num = np.zeros(len(yi))
    for i in range(len(xi)):
        denom[i] = kernel(x, xi[i], var)
    for i in range(len(xi)):
        num[i] = kernel(x, xi[i], var)*y[i]
    return np.sum(num)/np.sum(denom)
```

```python
model = np.zeros(len(x))
for i in range(len(x)):
    model[i] = nadaraya_watson_estimator(x[i],x,y,5) #tune var for making the model better
```

```python
plt.scatter(x,y, label = 'Actual Data')
plt.plot(x,model, color = 'r', label = 'Model')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
