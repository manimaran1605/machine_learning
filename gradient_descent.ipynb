```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

# Generate X values
X = np.arange(1, 20)  # [1, 2, ..., 10]

# Perfect linear relationship (Y = 2X)
Y_perfect = 2 * X

# Linear relationship with some noise
noise = np.random.normal(0, 0.5, size=X.shape)
y = 2 * X + noise
```

```python
plt.plot(X, y)
```

```python
def updated_w_and_b(X,y,w,b, alpha):
    dl_dw = 0
    dl_db = 0
    N = len(y)
    for i in range(N):
        dl_dw += -2 * X[i] * (y[i] - (w*X[i] + b))
        dl_db += -2*(y[i]- (w*X[i] + b))

    w = w- (1/float(N))*dl_dw*alpha
    b = b - (1/float(N))*dl_db*alpha

    return w, b
```

```python
def train(X, y, w, b, alpha, epochs):
    loss = []
    for e in range(epochs):
        w, b = updated_w_and_b(X,y,w,b, alpha)

        if e % 400 == 0:
            temp =  avg_loss(X, y, w, b)
            print("epoch", 0, "loss: ", temp)
            loss.append(temp)
    return w, b, loss
```

```python
def avg_loss(X,y,w,b):
    N = len(y)
    total_error = 0.0
    for i in range(N):
        total_error += (y[i] - (w*X[i] + b))**2
    return total_error /float(N)  
```

```python
w_opt, b_opt, loss_fun = train(X,y,0,0,0.001,1500)
```
