```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
logit = np.array([0.5, 0.1, 1.6, 0.8, 1.1, 0.2])
def softmax(x): #converts logits to probabilities
    softmax_values = np.zeros(len(logit))
    norm = 0
    for i in range(len(logit)):
        norm += np.exp(logit[i])
    for i in range(len(logit)):
        softmax_values[i] = np.exp(logit[i])/norm
    return softmax_values

softmax_values = softmax(logit)
plt.plot(logit, label = 'logit')
plt.plot(softmax_values, label = 'softmax')
plt.legend()
plt.show()
```
